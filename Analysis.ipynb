# Analysis of child illiteracy rates over time in several countries

This project will look at datasets about child illiteracy and outcomes in several countries

## Data initialisation and initial exploration

Summary:

3 datasets are being used, the one one about interventions does not seem to be related very well to the other two datasets. The dataset about illiteracy in countries and children out of school could be merged on the country code and year and then analysed together.

```python
# import the relevant modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
print("Libraries imported successfully!")
```

```python
#reading in our data

# interventions
interventions_df = pd.read_csv("intervention_programs.csv")
interventions_df['name'] = 'Interventions'

# out of school children
not_schooled_df = pd.read_csv("out_of_school_children.csv")
not_schooled_df['name'] = 'Out of school children'

# literacy rates
country_literacy_df = pd.read_csv("child_illiteracy_by_country.csv")
country_literacy_df['name'] = 'Illiteracy rates by country'

# if this runs successfully
print("Datasets successfully imported")
```

```python
# getting the shape and descriptive statistics for each dataset

# a list to hold the datasets for quicker exploration
datasets = [interventions_df, country_literacy_df, not_schooled_df]

# for each dataset, get the shape and summary stats and columns
for dataset in datasets:
    print(f"Details for {dataset['name'][0]} dataset")
    print(f"Info: {dataset.info}")
    print(f"Shape: {dataset.shape}")
    print(f"Columns: {dataset.columns}")
    print("Summary statistics: ")
    print(dataset.describe())
    print()
    print()
    print()
    print()
```

## Merging the datasets

Summary: 

```python
children_illiteracy_df = pd.merge(country_literacy_df, not_schooled_df, how='outer', left_on=['country', 'year'], right_on=['country', 'year'])
```
